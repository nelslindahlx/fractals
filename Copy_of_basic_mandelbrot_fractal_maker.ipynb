```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for i in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return max_iter

def create_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter):
    image = np.zeros((height, width))

    for row in range(height):
        for col in range(width):
            x = np.interp(col, [0, width], [x_min, x_max])
            y = np.interp(row, [0, height], [y_min, y_max])
            c = x + y * 1j

            iteration = mandelbrot(c, max_iter)
            color = (iteration / max_iter) ** 0.2  # Adjust color intensity

            image[row, col] = color

    return image

# Set the size and properties of the fractal
width = 800
height = 600
x_min = -2.5
x_max = 1
y_min = -1.5
y_max = 1.5
max_iter = 1000

# Generate the fractal image
fractal = create_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter)

# Display the fractal image
plt.figure(figsize=(10, 8))
plt.imshow(fractal, cmap='hot', extent=(x_min, x_max, y_min, y_max))
plt.colorbar(label='Iteration')
plt.title('Mandelbrot Set')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```
