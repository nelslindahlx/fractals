```python
# prompt: make a mandelbrot fractal

import numpy as np
import matplotlib.pyplot as plt

# Define the range of complex numbers to be plotted
x_min = -2.0
x_max = 1.0
y_min = -1.5
y_max = 1.5

# Define the number of pixels in the x and y directions
pixel_x = 1024
pixel_y = 1024

# Create an array to store the pixel values
image = np.zeros((pixel_y, pixel_x))

# Calculate the mandelbrot set for each pixel
for i in range(pixel_y):
    for j in range(pixel_x):
        c = complex(x_min + (x_max - x_min) * j / pixel_x,
                    y_min + (y_max - y_min) * i / pixel_y)
        z = c
        iteration = 0
        while abs(z) <= 2 and iteration < 255:
            z = z ** 2 + c
            iteration += 1
        image[i, j] = iteration

# Plot the image
plt.imshow(image, cmap='hot', interpolation='nearest')
plt.show()
```
