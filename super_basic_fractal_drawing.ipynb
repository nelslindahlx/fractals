Prompt: write python code to make a fractal for google colab notebook

```python
import matplotlib.pyplot as plt

def draw_fractal(x, y, size, n):
  """
  Draws a fractal at the point (x, y) with size s and n iterations.

  Args:
    x: The x-coordinate of the point.
    y: The y-coordinate of the point.
    size: The size of the fractal.
    n: The number of iterations.
  """

  if n == 0:
    return

  plt.plot([x, x + size * 0.5], [y, y + size * 0.866], 'b')
  draw_fractal(x + size * 0.5, y + size * 0.866, size / 2, n - 1)
  draw_fractal(x - size * 0.5, y + size * 0.866, size / 2, n - 1)

def main():
  """
  The main function.
  """

  draw_fractal(0, 0, 100, 6)

  plt.show()

if __name__ == "__main__":
  main()
```

Prompt: write python code to make a random fractal for google colab notebook

```python
import matplotlib.pyplot as plt
import random

def draw_fractal(x, y, size, n):
  """
  Draws a random fractal at the point (x, y) with size s and n iterations.

  Args:
    x: The x-coordinate of the point.
    y: The y-coordinate of the point.
    size: The size of the fractal.
    n: The number of iterations.
  """

  if n == 0:
    return

  x1 = x + random.randint(-size, size)
  y1 = y + random.randint(-size, size)
  x2 = x + random.randint(-size, size)
  y2 = y + random.randint(-size, size)

  plt.plot([x, x1], [y, y1], 'b')
  plt.plot([x, x2], [y, y2], 'b')
  draw_fractal(x1, y1, size / 2, n - 1)
  draw_fractal(x2, y2, size / 2, n - 1)

def main():
  """
  The main function.
  """

  draw_fractal(0, 0, 100, 6)

  plt.show()

if __name__ == "__main__":
  main()
```
