Prompt for Bard: write python code to make a fractal for google colab notebook

```python
import matplotlib.pyplot as plt

def draw_fractal(x, y, size, n):
  """
  Draws a fractal at the point (x, y) with size s and n iterations.

  Args:
    x: The x-coordinate of the point.
    y: The y-coordinate of the point.
    size: The size of the fractal.
    n: The number of iterations.
  """

  if n == 0:
    return

  plt.plot([x, x + size * 0.5], [y, y + size * 0.866], 'b')
  draw_fractal(x + size * 0.5, y + size * 0.866, size / 2, n - 1)
  draw_fractal(x - size * 0.5, y + size * 0.866, size / 2, n - 1)

def main():
  """
  The main function.
  """

  draw_fractal(0, 0, 100, 6)

  plt.show()

if __name__ == "__main__":
  main()
```

Prompt for Bard: write python code to make a random fractal for google colab notebook

```python
import matplotlib.pyplot as plt
import random

def draw_fractal(x, y, size, n):
  """
  Draws a random fractal at the point (x, y) with size s and n iterations.

  Args:
    x: The x-coordinate of the point.
    y: The y-coordinate of the point.
    size: The size of the fractal.
    n: The number of iterations.
  """

  if n == 0:
    return

  x1 = x + random.randint(-size, size)
  y1 = y + random.randint(-size, size)
  x2 = x + random.randint(-size, size)
  y2 = y + random.randint(-size, size)

  plt.plot([x, x1], [y, y1], 'b')
  plt.plot([x, x2], [y, y2], 'b')
  draw_fractal(x1, y1, size / 2, n - 1)
  draw_fractal(x2, y2, size / 2, n - 1)

def main():
  """
  The main function.
  """

  draw_fractal(0, 0, 100, 6)

  plt.show()

if __name__ == "__main__":
  main()
```

Prompt for ChatGPT: write python code to make a fractal for google colab notebook

```python
import numpy as np
import matplotlib.pyplot as plt

def fractal(x, y, size):
    if size < 3:
        return
    else:
        # Draw the outer square
        plt.plot([x, x+size, x+size, x, x], [y, y, y+size, y+size, y], 'k')
        
        # Draw the inner square
        inner_size = size / 3
        fractal(x + inner_size, y + inner_size, inner_size)
        
        # Draw the other eight inner squares
        for i in range(3):
            for j in range(3):
                if i == 1 and j == 1:
                    continue
                fractal(x + i * inner_size, y + j * inner_size, inner_size)

# Set the figure size and dpi
plt.figure(figsize=(6, 6), dpi=80)

# Call the fractal function
fractal(0, 0, 81)

# Adjust the plot limits
plt.xlim(-10, 90)
plt.ylim(-10, 90)

# Remove the axis ticks
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def fractal(x, y, size, randomness):
    if size < 3:
        return
    else:
        # Draw the outer square
        plt.plot([x, x+size, x+size, x, x], [y, y, y+size, y+size, y], 'k')
        
        # Draw the inner square with random offset
        inner_size = size / 3
        offset = random.uniform(-randomness, randomness) * inner_size
        fractal(x + inner_size + offset, y + inner_size + offset, inner_size, randomness)
        
        # Draw the other eight inner squares with random offset
        for i in range(3):
            for j in range(3):
                if i == 1 and j == 1:
                    continue
                offset = random.uniform(-randomness, randomness) * inner_size
                fractal(x + i * inner_size + offset, y + j * inner_size + offset, inner_size, randomness)

# Set the figure size and dpi
plt.figure(figsize=(6, 6), dpi=80)

# Call the fractal function with random parameters
randomness = 0.4
fractal(0, 0, 81, randomness)

# Adjust the plot limits
plt.xlim(-10, 90)
plt.ylim(-10, 90)

# Remove the axis ticks
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```
